```python
plt.plot([1,2,3],[1,4,9])
plt.show()
```

```python
k=['kanon','kpol','ayon','riad']
age=[25,17,13,7]
color=[5,2,1,1]
plt.plot(k,age,label='first')
plt.plot(k,color,label='second')
plt.legend()
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Name vs age')
plt.show()
```

```python
plt.bar(k,age,label='age_bar',align='center')
plt.bar(k,color,label='color_scale_bar')
plt.legend()
plt.show()
```
