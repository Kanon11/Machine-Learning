# How do I filter rows of a pandas DataFrame by column value?

```python
booleans = []
```

```python
for amount in df.Sales :
    if amount>=3000:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
how_sales=pd.Series(booleans)
```

```python
df[how_sales]
```
