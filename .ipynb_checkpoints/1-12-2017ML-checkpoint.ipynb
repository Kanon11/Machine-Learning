```python
import pandas as pd
```

```python
data=pd.read_csv('Advertising.csv',index_col=0)
```

```python
data.head()
```

```python
data.shape
```

```python
import seaborn as sns
%matplotlib inline
```

```python
# visualization the relationship between features and respons using scaterplot
sns.pairplot(data,x_vars=['TV','radio','newspaper'],y_vars='sales',size=5,aspect=0.9,kind='reg')
```

```python
type(data)
```

```python
X=data[['TV','radio','newspaper']]
```

```python
X.values
```

```python
y=data['sales']
```

```python
y.values
```

```python
print(X.shape)
print(y.shape)
```

# Spliting X & y into Training and Test data

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
```

# Intercepting model coefficients


```python
print(LR.intercept_)
```

```python
print(LR.coef_)
```

# Making prediction

```python
ture=[100,50,30,20]# testing
pred=[90,50,50,30]
from sklearn import metrics
print(metrics.mean_squared_error(ture,pred))
```

# computing RMSE prediction for our prediction

```python
import numpy as np
print (np.sqrt(metrics.mean_squared_error(y_pred,y_test)))
```
