```python
def get_data(symbols,dates):
   # start_date='09-05-2014'
    #end_date='09-05-2016'
    #dates=pd.date_range(start_date,end_date)
    
    #print(dates.values)
    # create an empty dataframe
    #df1=pd.DataFrame(index=dates)
    
    # read SPY data on a temporary dataframe
    #dfspy=pd.read_csv('data/SPY.csv',index_col='Date',usecols=['Date','Adj Close'],na_values='nan',parse_dates=True)
    #dfspy=dfspy.rename(columns={'Adj Close':'SPY'})
    
    #df1=df1.join(dfspy,how='inner')
    
    #for joining more data
    #symbols=['GOOG','IBM','GLD']
    
    df1=pd.DataFrame(index=dates)
    
    if 'SPY' not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, 'SPY')
    
    for symbol in symbols:
        df_temp=pd.read_csv('data/{}.csv'.format(symbol),index_col='Date',usecols=['Date','Adj Close'],parse_dates=True)
        df_temp=df_temp.rename(columns={'Adj Close':symbol})
        df1=df1.join(df_temp)
    
    df1=df1.dropna()
    return df1
    
    
```

```python
def data_plot(df, title='Stock Prices', xlabel="Date", ylabel="Price"):
    ax1=df.plot(title=title,fontsize=14,figsize=(17,8))
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)
    #plt.xlabel('Data')
    #plt.ylabel('Price')
    plt.show()
```

```python
def get_rolling_mean(values, window):
    """Return rolling mean of given values, using specified window size."""
    return pd.rolling_mean(values, window=window)


def get_rolling_std(values, window):
    """Return rolling standard deviation of given values, using specified window size."""
    # TODO: Compute and return rolling standard deviation
    return pd.rolling_std(values,window=window)


def get_bollinger_bands(rm, rstd):
    """Return upper and lower Bollinger Bands."""
    # TODO: Compute upper_band and lower_band
    upper_band = rm + rstd * 2
    lower_band = rm - rstd * 2

    return upper_band, lower_band
```

```python
def daily_return_panda(df):
    daily_returns=(df/df.shift(1))-1
    #daily_returns.ix[0,:]=0
    return daily_returns
```

```python
def portfolio():
    symbols=['SPY','XLB','GOOG','GLD']
    dates=pd.date_range('2009-01-01','2011-12-31')
    df=get_data(symbols,dates)
    allocs=[0.4,0.4,0.1,0.1]
    ndf=df_normalization(df)
    allocated=ndf*allocs
    start_val=1000000
    pos_vals=start_val*allocated
    port_vals=pos_vals.sum(axis=1)
    daily_returns_of_port_vals=daily_return_panda(port_vals)
    daily_returns_of_port_vals=daily_returns_of_port_vals[1:]
    return daily_returns_of_port_vals
```
