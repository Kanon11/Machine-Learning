# Titanic: Machine Learning from Disaster ( for Kaggle submission)

```python
import pandas as pd
import numpy as np
```

```python
dataframe=pd.read_csv("train.csv")
```

```python
dataframe.head(7)
```

```python
feature_cols=['Pclass','Age','Sex','SibSp','Parch']
```

```python
X=dataframe.loc[:,feature_cols]
```

```python
y=dataframe.Survived
```

```python
X.shape
```

```python
y.shape
```

```python
map_sex={"male":0,"female":1}
```

```python
X['Sex']=X['Sex'].map(map_sex)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

import matplotlib.pyplot as plt
import seaborn as sns

dt=DecisionTreeClassifier()
rf=RandomForestClassifier(n_estimators=200)
LG=LogisticRegression()
```

```python
X.isnull().sum()
```

```python
age_mean=X.Age[X.Age.notnull()].mean()
```

```python
age_mean
```

```python
X.Age.fillna(value=age_mean,inplace=True)
```

```python
X.head(10)
```

```python
LG.fit(X,y)
dt.fit(X,y)
rf.fit(X,y)
```

```python
dataframe_test=pd.read_csv('test.csv')
```

```python
X_test=dataframe_test.loc[:,feature_cols]
```

```python
X_test
```

```python
X_test.Sex=X_test.Sex.map(map_sex)
```

```python
test_age_mean=X_test.Age[X_test.Age.notnull()].mean()
```

```python
test_age_mean
```

```python
X_test.Age.fillna(value=test_age_mean,inplace=True)
```

```python
y_predict_class=LG.predict(X_test)
y_predict_dt=dt.predict(X_test)
y_predict_rf=rf.predict(X_test)
```

```python
y_predict_class.sum()
```

```python
print(y_predict_dt.sum())
print(y_predict_rf.sum())
```
