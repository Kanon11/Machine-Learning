```python
import numpy as np
import pandas as pd
```

```python
dataframe=pd.read_csv("pepsi.csv",index_col=0)
```

```python
type(dataframe)
```

```python
dataframe.shape
```

```python
import seaborn as sns
%matplotlib inline
```

```python
feature_col=['Temperature','Class-ON/OFF','Festival Priority Value']
respons_col='Pepsi Unit Sales'
# viewing the relationship between features column and respons column
sns.pairplot(dataframe,x_vars=feature_col,y_vars=respons_col,size=5,aspect=0.9,kind='reg')
```

```python
X=dataframe[feature_col]
y=dataframe[respons_col]
```

```python
X.values
```

```python
y.values
```

```python
print(X.shape)
```

```python
print(y.shape)
```

# Using train test split to X and y

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Modeling our data using Linear Regrassion

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
y_predict=LR.predict(X_test)
```

```python
y_predict
```

# Intercepting model coefficients

```python
A=LR.intercept_
print(LR.intercept_)
```

```python
x=LR.coef_
print(LR.coef_)
```

```python
x
```

```python
x1=int(input("enter your Temperature"))
```

```python
x2=int(input("enter class ON/OFF, 1 for ON 0 for OFF"))
```

```python
x3=int(input("enter your festival priority (higher the better)"))
```

```python
y=A+x[0]*x1+x[1]*x2+x[2]*x3
print(round(y)," unit Pepsi will be sold")
```

# computing RMSE prediction for our model

from sklearn import metrics

```python

print(np.sqrt(metrics.mean_squared_error(y_predict,y_test)))
```

# Modeling our data in Logestic Regrassion 

```python
from sklearn.linear_model import LogisticRegression
```

```python
LG=LogisticRegression()
```

```python
LG.fit(X_train,y_train)
```

```python
y_pre_for_LG=LG.predict(X_test)
```

```python
y_pre_for_LG
```

```python
print(metrics.accuracy_score(y_test,y_pre_for_LG))
```

# comparing our Model with y_test with RMSE

```python
print(np.sqrt(metrics.mean_squared_error(y_pre_for_LG,y_test)))
```
